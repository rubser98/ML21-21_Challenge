```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

df = pd.read_csv('OnlineNewsPopularity_cleaned.csv', sep=';',low_memory=False, na_values='n.a.')
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp.fit(df)
dataset = pd.DataFrame(imp.transform(df),columns=df.columns)
dataset.to_csv('OnlineNewsPopularity_cleanedv2.csv', sep=';')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, auc, roc_curve
from sklearn.metrics import plot_confusion_matrix, confusion_matrix
from sklearn.model_selection import cross_val_predict

import warnings
warnings.filterwarnings('ignore')
excluded_features = ['url',' timedelta',' shares','shares_class',' weekday_is_monday',' weekday_is_tuesday'
                     ,' weekday_is_wednesday',' weekday_is_thursday',' weekday_is_friday',' weekday_is_saturday'
                     ,' weekday_is_sunday',' data_channel_is_bus',' n_non_stop_words',' n_tokens_content'
                     ,' average_token_length',' data_channel_is_lifestyle',' global_rate_negative_words'
                     ,' min_negative_polarity']

X = dataset.drop(excluded_features, axis=1)
Y = dataset['shares_class']
#scaling feature values
X_scaled = preprocessing.MinMaxScaler().fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.30, random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn import svm

kernel = ['rbf', 'linear', 'poly', 'sigmoid']
C=[0.1, 1, 10, 100, 1000]

tuned_parameters = dict(kernel=kernel,C=C)
print(tuned_parameters)
scores = ['precision', 'recall']

for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()

    clf = GridSearchCV(
       svm.SVC() , tuned_parameters, scoring='%s_macro' % score
    )
    clf.fit(X_train, y_train)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
              % (mean, std * 2, params))
    print()

    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_test, y_pred))
    print()
```

```python
The dataset collects 39.000 articles of Mashable news website during 2 years period. The dataset has some noise: I excluded 
the row which has missing value in the target and the missing values for num_imgs and num_videos are substituted by most 
frequent values. The aim of the challenge is to predict the shares of articles which is a continuous variable.In the histogram that
plots shares distribution, we can see that there is a skewness. So I performed SMOGN (SMOte for regression with Gaussian Noise) 
to resampling data adjusting the skewed distribution. SMOGN analyzes data and detects which examples are more and less relevant 
to the distribution according to a relevance score criteria. Then, SMOGN generates syntethic examples from most relevant ones 
and undersamples from less relevant group.
In order to identify most relevant features I have computed SelectKBest algorithm which uses Mutual Information. Most revelant 
feature are the ones which linked to keywords and the ones based on Latent Dirichlet allocation(LDA) that identify closeness to 
the five most relevant topic in Mashable. The less relevant features are the ones related to publication day, average words 
length and some related to sentiment polarity. 
To semplify the task, I discretized the target into five classes (i.e. 0: not popular up to 4: very popular). Discretization is
performed for either resampled and not resampled dataset. As we can see on the class distribution histograms, SMOGN pays more 
attention to resample instances that lie in the class 1. 
The chosen algorithm to classify data are: Support Vector Machines(SVM), Decision Tree (DT), Multilayer Perceptron (MLP) and 
for ensemble method, Random Forest (RF). Since that algorithm like SVM and MLP are sensitive to feature scale, I performed 
MinMax Scaling to the dataset allowing better performance for that algorithms.
I removed less relevant features to reduce input dimension allowing to reduce computational cost and noise. First, I start
training and testing models for original dataset. Before training the models I performed GridSearch to detect the best 
hyperparameters. GridSearch fits and tests the models with each combination of the hyperparameter indicated in input. 
The hyperparameters chosen by GridSearch are the following:
    - SVM: radial basis function (RBF) kernel and C = 1
    - DT: best splitting, maximum depth = 6 and gini criterion
    - MLP: tanh as activation function, alpha= 0.05, hidden layer sizes=(50, 50, 50), constant learning rate, adam as optimizer
    - RF:  entropy criterion, max_features is square root of number of features ,number of trees=300
        
Tests are performed with 5-fold cross validation and metrics used to evaluate models are accuracy, precision and recall:
    - Accuracy = number of correctly classified instances / number of instances
    - Precision = True Positive / True Positive  + False Positive
    - Recall = True Positive / True Positive + False Negative

The results are the following:
              Accuracy Precision Recall
    - SVM:     0.30     0.30      0.30
    - DT:      0.30     0.30      0.30
    - MLP:     0.30     0.30      0.30
    - RF:      0.31     0.31      0.31

Classifiers have very similar performance. The results below show performance over the resampled dataset:
                  Accuracy Precision Recall
    - RBF-SVM:     0.74     0.74      0.74
    - linear-SVM:  0.83     0.83      0.83
    - DT:          0.88     0.88      0.88
    - MLP:         0.84     0.84      0.84
    - RF:          0.87     0.87      0.87
        
Results show that with non-resampled dataset models have similar performance, around 30%. Performance considerably increases
with resampled dataset, from 74%, with RBF-SVM, up to 88% of DT. We can also observe that, on resampled dataset, SVM works 
better with a linear decision boundary instead with radial basis.
```
